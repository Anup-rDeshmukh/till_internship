```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
marks_of_student=np.random.randint(0,100,(50))
marks_of_student
```

```python
plt.hist(marks_of_student)
plt.show
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(8,8))
plt.hist(marks_of_student,bins=bins,edgecolor="black", linestyle="--",color="red")

plt.xticks(np.arange(0,100,5))

plt.show()
```

```python
plt.hist(marks_of_student,bins=bins,edgecolor="black", linestyle="--",color="red", rwidth=0.6)
plt.show()
```

```python
mark_of_student1=np.random.randint(0,100,(50))
mark_of_student2=np.random.randint(0,100,(50))
plt.figure(figsize=(8,9))
bins=np.arange(0,100,5)
plt.hist([mark_of_student1,mark_of_student2],bins=bins, color=['red','blue'],rwidth=0.6)
plt.xticks(np.arange(0,100,5))
plt.title("class_A,B_comparison")

plt.show()
```
