```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

classes=['maths','english','physics','chemistry','biology']
marks=[67,87,65,56,54]

plt.pie(marks,labels=classes, autopct='%1.1f%%')
plt.show
```

```python
explode_values=[0.1,0,0,0,0]
plt.pie(marks,labels=classes, autopct='%1.1f%%' , explode=explode_values)
plt.show()
```

```python
textprops={'fontsize':14,'color':'k'}
plt.figure(figsize=(8,8))
wedgeprops={'linewidth':3,'linestyle':'--','edgecolor':'white'}
plt.pie(marks,labels=classes,autopct='%1.1f%%', textprops=textprops, wedgeprops=wedgeprops, explode=explode_values)

plt.legend()
plt.show()
```
